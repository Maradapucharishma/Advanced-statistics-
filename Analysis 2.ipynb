### 1) Can you tell me No of Vehicles by Brand Available on ebay for sale with the help of visualization

```python
df['brand'].value_counts().plot(kind='bar')
```

* Highest sales of cars on the basis of brand is volkswagen

###  2) What is the Average price for vehicles based on the type of vehicle as well as on the type of gearbox.Explain me with both numerical and visualization analysis

```python
pd.crosstab(index =df['vehicleType'],columns =df["gearbox"], values = df['price'],aggfunc='mean' ).plot.bar()
```

* The average price of vehicles on the basis of type of vehicle and type of gearbox , SUV have the highest average price
* keinwagen has the lowest average price
